```python
import numpy as np 
import matplotlib.pyplot as plt
#python 3.10.4
```

```python
# Generate data using a dummy function 
X = np.arange(10)
Y = (X-5)**2 + 3 

print(X,Y)
```

```python
# Visualise 
import seaborn
plt.style.use("seaborn-v0_8-deep")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
# Gradient Descent 
# init x with any random value 
x = 1
lr = 0.1 # learning rate

plt.plot(X,Y)
for i in range(30):
    grad = 2*(x-5)
    x = x - lr * grad 
    y = (x-5)**2 + 3 
    plt.scatter(x,y)
    print(x)
plt.grid(True)
plt.show()
```
