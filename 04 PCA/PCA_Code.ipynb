```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
breast_cancer = load_breast_cancer()
#breast_cancer
```

```python
X = breast_cancer['data']
print(X)
```

```python
X.shape
```

#### Step 1

```python
mu = X.mean(axis=0)
# mean centering
X_ = (X - mu)
X_.shape
```

#### Step 2 - Calculate Covariance Matrix


```python
S = np.dot(X_.T, X_)
S.shape
```

#### Step 3 : Calculate Eigen vector and Eigen values

```python
lambdas, vectors = np.linalg.eig(S)
#vectors
#lambdas
```

```python
lambdas.shape
```

```python
lambdas[:2]
```

```python
vectors.shape
```

```python
## Eigen Vectors
vectors
```

#### Step 4 : Converting 30D to 2D dimension vector

```python
V = vectors[:, :2]
V.shape
```

#### Step 5 Projection

```python
X_.shape
```

```python
V.shape
```

```python
X_new = np.dot(X_, V)
X_new.shape
```

```python
X_new
```

```python
# Eigen Values
lambdas
```

```python
total = sum(lambdas)
total
```

```python
explained_variance = np.round(lambdas/total, 3)
explained_variance
```

```python
explained_variance[:2].sum()
```

```python
explained_variance
```

```python
cummulative_variance =  np.cumsum(explained_variance)
cummulative_variance
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot( range(1, len(cummulative_variance)+1) , cummulative_variance, marker="o")
#plt.plot(cummulative_variance, marker="o")
plt.xlabel("Number of components")
plt.ylabel("Variance Explained")
plt.grid("on")
```
