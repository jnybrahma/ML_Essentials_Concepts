```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(legacy='1.25')

import warnings
warnings.filterwarnings('ignore')

from sklearn.datasets import make_moons, make_circles, make_classification
```

```python
X_moons, Y_moons = make_moons(n_samples=2000, noise=0.15, random_state=1)
X_circle, Y_circle = make_circles(n_samples=2000, noise=0.1, random_state=1, factor=0.5)
X_class, Y_class = make_classification(n_samples=2000, n_features=2, n_redundant=0, random_state=1)
```

```python
plt.scatter(X_moons[:,0],X_moons[:,1], c=Y_moons)
plt.show()
```

```python
plt.scatter(X_circle[:,0],X_circle[:,1], c=Y_circle)
plt.show()
```

```python
plt.scatter(X_class[:,0],X_class[:,1], c=Y_class)
plt.show()
```

```python
def load_dataset(dataset):
    if dataset=='moons':
        X, Y = make_moons(n_samples=10000, noise=0.15, random_state=1)
    elif dataset=='circles':
        X, Y = make_circles(n_samples=10000, noise=0.1, random_state=1, factor=0.5)
    elif dataset=='classification':
        X, Y = make_classification(n_samples=10000, n_features=2, n_redundant=0, random_state=2)

    return X,Y
```

```python
dataset = 'moons'
X,Y = load_dataset(dataset)
print(X.shape, Y.shape)
```

```python
plt.scatter(X[:, 0], X[:, 1], c = Y)
plt.show()
```

# Model Building

```python
from keras import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add( Dense(10, activation='relu', input_shape = (2,) )  )  # 2 neuron, 2 input_shape is given for the first hidden layer.
model.add( Dense(5, activation='relu') )   # add 2nd layer has 5 neuron
model.add( Dense(1, activation='sigmoid') ) # Output layer


model.compile(optimizer='adam',loss='binary_crossentropy',   metrics=['accuracy'])
model.summary()
```

```python
model.fit(X, Y, batch_size = 64, epochs=50, validation_split=0.1)
```
