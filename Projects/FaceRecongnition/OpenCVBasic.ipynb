## Object Detection using Haarcascades Classifier

```python
img = cv2.imread("GroupImage.jpg")
img.shape
```

```python
# Show this image
plt.imshow(img)
```

```python
newImg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Show this image
plt.imshow(newImg)
plt.axis("off")
plt.show()
```

```python
model = cv2.CascadeClassifier("haarcascade_frontalface_alt.xml")
```

```python
faces = model.detectMultiScale(newImg,1.3,1)
```

```python
print(faces.shape)
faces
```

```python
for f in faces:
    x,y,w,h = f
    cv2.rectangle(newImg,(x,y),(x+w, y+h),(0,255,0),2)

plt.imshow(newImg)
plt.show()
```
